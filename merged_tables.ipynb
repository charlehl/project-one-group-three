```python
import pandas as pd
```

```python
table_2013 = "Resources/top10table2013.csv"
table_2014 = "Resources/top10table2014.csv"
table_2015 = "Resources/top10table2015.csv"
table_2016 = "Resources/top10table2016.csv"
table_2017 = "Resources/top10table2017.csv"
```

```python
table_2013_df = pd.read_csv(table_2013)
table_2014_df = pd.read_csv(table_2014)
table_2015_df = pd.read_csv(table_2015)
table_2016_df = pd.read_csv(table_2016)
table_2017_df = pd.read_csv(table_2017)
```

```python
# merged tables dataframes by destination city(which in reality represents the months)
first_merge = pd.merge(table_2013_df,table_2014_df, on="Destination City")
second_merge = pd.merge(first_merge, table_2015_df, on="Destination City")
third_merge = pd.merge(second_merge, table_2016_df, on="Destination City")
merged_all_years = pd.merge(third_merge, table_2017_df, on="Destination City")
merged_all_years_rename = merged_all_years.rename(columns={"Destination City":"Index"})
```

```python
merged_set_index = merged_all_years_rename.set_index("Index")
merged_transposed = merged_set_index.T
merged_reset_index = merged_transposed.reset_index()
merged_sorted = merged_reset_index.sort_values("index")
merged_renamed = merged_sorted.rename(columns={"index":"Destination City",1:"January",2:"February",3:"March",
                                              4:"April",5:"May",6:"June",7:"July", 8:"August",9:"September",
                                              10:"October",11:"November",12:"December"})
#last_index_reset = merged_renamed.reset_index()
merged_organized = merged_renamed[["Destination City","January","February","March","April","May","June",
                                  "July","August","September","October","November","December"]]
```

```python
merged_organized.reset_index(drop=True)
```

```python
merged_organized.to_csv("Resources/all_cities_years.csv")
```
