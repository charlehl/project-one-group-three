* create dataframe using the csv file "carriers_2017.csv"
* narrow it down to LAX and JFK
* get a count of all travel destinations for each city

```python
# import dependencies
import pandas as pd
```

```python
# open csv file
file = "carriers_2017.csv"

carriers_df = pd.read_csv(file)
```

```python
carriers_df.head()
```

```python
# create a new dataframe wiht only origin LAX and JFK
# user loc to filter by ORIGIN
lax_carriers = carriers_df.loc[carriers_df["ORIGIN"] == "LAX",:]
# rename columns with DEST_CITY_NAME and SEATS for the final destinations dataframe
lax_carriers = lax_carriers.rename(columns={"DEST_CITY_NAME":"Destination City","CARRIER_NAME":"Flight Count"})
jfk_carriers = carriers_df.loc[carriers_df["ORIGIN"] == "JFK",:]
# rename columns with DEST_CITY_NAME and SEATS for the final destinations dataframe
jfk_carriers = jfk_carriers.rename(columns={"DEST_CITY_NAME":"Destination City","CARRIER_NAME":"Flight Count"})
```

```python
# group new LAX dataframe by Destination City
grouped_lax = lax_carriers.groupby("Destination City").count()
grouped_lax = grouped_lax["Flight Count"]
lax_sorted = grouped_lax.sort_values(ascending=False)
lax_destinations = pd.DataFrame(lax_sorted)
lax_destinations.head()
```

```python
# group new JFK dataframe by Destination City
grouped_jfk = jfk_carriers.groupby("Destination City").count()
grouped_jfk = grouped_jfk["Flight Count"]
jfk_sorted = grouped_jfk.sort_values(ascending=False)
jfk_destinations = pd.DataFrame(jfk_sorted)
jfk_destinations.head()
```
